```python
from ultralytics import YOLO
import cv2

# =========================
# 1Ô∏è‚É£ Load Model
# =========================
model = YOLO("/content/runs/detect/train/weights/best.pt")

# =========================
# 2Ô∏è‚É£ Load Video
# =========================
video_path = "/content/Problem Statement Scenario1.mp4"
cap = cv2.VideoCapture(video_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Save output video
out = cv2.VideoWriter(
    "output_counted.mp4",
    cv2.VideoWriter_fourcc(*'mp4v'),
    fps,
    (frame_width, frame_height)
)

# =========================
# 3Ô∏è‚É£ Settings
# =========================
line_x = int(frame_width * 0.5)   # adjust if needed
count = 0

counted_ids = set()
last_side = {}   # store previous side of each track

# =========================
# 4Ô∏è‚É£ Process Video
# =========================
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(frame, persist=True, conf=0.4)

    if results[0].boxes.id is not None:
        boxes = results[0].boxes.xyxy.cpu().numpy()
        ids = results[0].boxes.id.cpu().numpy()

        for box, track_id in zip(boxes, ids):
            x1, y1, x2, y2 = box
            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)

            # Draw bounding box
            cv2.rectangle(frame, (int(x1), int(y1)),
                          (int(x2), int(y2)), (0,255,0), 2)

            # Draw center dot
            cv2.circle(frame, (cx, cy), 5, (255,0,0), -1)

            # Show ID
            cv2.putText(frame, f"ID:{int(track_id)}",
                        (int(x1), int(y1)-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6,
                        (0,255,0), 2)

            # Determine current side
            # True = right side, False = left side
            current_side = cx > line_x

            if track_id not in last_side:
                last_side[track_id] = current_side
            else:
                # RIGHT ‚Üí LEFT crossing
                if last_side[track_id] and not current_side:
                    if track_id not in counted_ids:
                        count += 1
                        counted_ids.add(track_id)

                last_side[track_id] = current_side

    # Draw vertical counting line
    cv2.line(frame, (line_x, 0),
             (line_x, frame_height),
             (0,0,255), 3)

    # Show count
    cv2.putText(frame, f"Count: {count}",
                (50, 60),
                cv2.FONT_HERSHEY_SIMPLEX,
                1.5, (0,0,255), 3)

    out.write(frame)

cap.release()
out.release()

print("Final Bag Count:", count)
print("Saved as output_counted.mp4")
```
